```python
from pyecharts.charts import Bar
from pyecharts.charts import Line
from pyecharts import options as opts

x_date = [
    '2020-02-09', '2020-02-10', '2020-02-11', '2020-02-12', '2020-02-13',
    '2020-02-14', '2020-02-15', '2020-02-16', '2020-02-17', '2020-02-18'
]

y_huangshi = [760, 805, 835, 874, 911, 943, 980, 988, 988, 985]

y_ezhou = [639, 725, 790, 861, 1065, 1125, 1192, 1230, 1274, 1339]
```

```python
bar=Bar()

bar.add_xaxis(x_date)

bar.add_yaxis('黄石',y_huangshi)

bar.set_global_opts(
    
    # 坐标轴配置，X和Y轴分别为xaxis_opts和yaxis_opts
    xaxis_opts=opts.AxisOpts(
        
        # 坐标轴名称
        name='日期',
        
        # 坐标轴标签配置
        # 还有is_show,font_size,font_family,font_weight等
        axislabel_opts=opts.LabelOpts(
            
            # 字体颜色
            color='red',
            
            # 标签与坐标间的距离
            margin=20,
            
            # 标签旋转
            rotate=30,
            
            # 文字样式：'normal'，'italic'，'oblique'
            font_style='italic',
            
            # 标签间隔，2表示每隔2个标签显示一个标签文字
            interval=2
        ),
        
        # 坐标轴刻度配置
        axistick_opts=opts.AxisTickOpts(
            
            # 是否显示刻度
            is_show=True,
            
            # 刻度线是否在内部
            is_inside=True,
            
            # 刻度线长度
            length=50,
            
            # 线段样式配置
            linestyle_opts=opts.LineStyleOpts(
                
                # 线段宽度
                width=5,
                
                # 透明度，0~1
                opacity=0.5,
                
                # 线段类型：'solid', 'dashed', 'dotted'
                type_='dotted',
                
                # 颜色，可以使用渐变色，具体参考文档
                color='blue'
            )
        ),
        
        # 坐标轴轴线配置
        axisline_opts=opts.AxisLineOpts(
            
            # 是否显示
            is_show=True,
            
            # 箭头：'none', 'arrow'，单个字符串表示两端都是同样的
            # ['none', 'arrow'] 数组表示两端不同的显示
            symbol=['none','arrow'],
            
            # 轴线线条样式配置，参考上方刻度线段配置
            linestyle_opts=opts.LineStyleOpts(
                width=5,
                color='green'
            )
        ),
        
        # 坐标轴提示框配置
        axispointer_opts=opts.AxisPointerOpts(
            
            # 是否显示
            is_show=True,
            
            # 提示框类型：'line' 直线指示器，'shadow' 阴影指示器，'none' 无指示器
            type_='line',
            
            # 提示框标签配置，参考上方坐标轴标签配置
            label=opts.LabelOpts(
                color='red',
                font_size=25
            ),
            
            # 轴线线条样式配置，参考上方刻度线段配置
            linestyle_opts=opts.LineStyleOpts(
                width=5,
                color='green',
                type_='dashed'
            )
        )
    ),
    
    yaxis_opts=opts.AxisOpts(
        name='确诊人数'
    )
)
bar.render_notebook()
```
